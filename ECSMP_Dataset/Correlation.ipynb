```python
import numpy as np 
import pandas as pd
from pandas import read_csv
import seaborn as sns
sns.set_theme
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('merge_drop_clean_dataset.csv', sep=',', header=None)
```

```python
df
```

```python
df.columns = df.iloc[-1] # define the header, which is the last row
```

```python
df = df[:-1] # drop the last row, which is header
```

```python
df['sds\t'] = df['sds\t'].str.strip()
```

```python
df
```

```python
df=df.rename(columns = {'sds\t':'sds'}) # reanme sds\t's column name as sds
```

```python
df
```

```python
# copy the data from the original dataset to create a new dataframe that includes 
# only the data I want to use to build the correlation matrix
new_df = df[['time_in_bed', 
                   'sleep_duration', 
                   'deep_sleep', 
                   'light_sleep', 
                   'rem', 
                   'wake', 'deep_sleep_onset', 
                   'sleep_efficiency', 
                   'psqi', 
                   'poms', 'expressive_suppression', 
                   'cognitive_reappraisal','sds']].copy()
```

```python
# change all columns' data type to float (original types are Object)
new_df = new_df.astype('float64')
```

```python
# create a correlation matrix using - correlation method used: Pearson coefficient of correlation
r = new_df.corr()
```

```python
r
```

```python
# plot the correlation matrix
plt.figure(figsize=(9,9))
sns.heatmap(r, annot=True, vmax=1, vmin=-1, center=0, cmap='vlag')
plt.savefig('correlation_all.png')
```

```python
# plot only the lower half correlation matrix
plt.figure(figsize=(9,9))
mask = np.triu(np.ones_like(r, dtype=bool))
sns.heatmap(r, annot=True, vmax=1, vmin=-1, center=0, cmap='vlag', mask=mask)
plt.savefig('correlation_half.png')
```

```python
# High degree Correlation: If the coefficient value lies between Â± 0.50 and Â± 1
new_r_high = r.unstack()
new_r_high = new_r_high[abs(new_r_high) >= 0.5]
new_r_high = new_r_high[abs(new_r_high) <= 1]
```

```python
# find only the correlation between Â± 0.50 and Â± 1
new_r_high
```

```python
# High degree Correlation except with itself: If the coefficient value lies between Â± 0.50 but NOT 1
new_r_high_not_itself = r.unstack()
new_r_high_not_itself = new_r_high_not_itself[abs(new_r_high_not_itself) >= 0.5]
new_r_high_not_itself = new_r_high_not_itself[abs(new_r_high_not_itself) != 1]
```

```python
new_r_high_not_itself
```

```python
# Moderate degree Correlation: If the value lies between Â± 0.30 and Â± 0.49
new_r_moderate = r.unstack()
new_r_moderate = new_r_moderate[abs(new_r_moderate) >= 0.3]
new_r_moderate = new_r_moderate[abs(new_r_moderate) <= 0.49]
```

```python
new_r_moderate
```

```python
# Low degree Correlation: If the value lies below Â± .29
new_r_low = r.unstack()
new_r_low = new_r_low[abs(new_r_low) <= 0.29]
new_r_low = new_r_low[abs(new_r_low) > 0]
```

```python
new_r_low
```

```python
# No Correlation: If the value equals to 0
new_r_no = r.unstack()
new_r_no = new_r_no[abs(new_r_no) == 0]
```
